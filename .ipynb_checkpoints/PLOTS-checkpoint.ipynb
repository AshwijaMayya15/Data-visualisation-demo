```python
import matplotlib.pyplot as plt

# Sample datasets
bar_data = {'Category': ['A', 'B', 'C', 'D'],
            'Value': [10, 5, 12, 8]}

line_data = {'Year': [2010, 2011, 2012, 2013],
             'Sales': [100, 150, 120, 180]}

pie_data = {'Category': ['A', 'B', 'C', 'D'],
            'Percentage': [30, 20, 25, 25]}

scatter_data = {'X': [1, 2, 3, 4, 5],
                'Y': [3, 5, 4, 6, 8]}

heatmap_data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

treemap_data = {'Category': ['A', 'B', 'C', 'D'],
                'Value': [10, 5, 12, 8]}

# Bar Chart
plt.figure()
plt.bar(bar_data['Category'], bar_data['Value'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart')

# Line Chart
plt.figure()
plt.plot(line_data['Year'], line_data['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Line Chart')

# Pie Chart
plt.figure()
plt.pie(pie_data['Percentage'], labels=pie_data['Category'], autopct='%1.1f%%')
plt.title('Pie Chart')

# Scatter Plot
plt.figure()
plt.scatter(scatter_data['X'], scatter_data['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')

# Heatmap
plt.figure()
plt.imshow(heatmap_data, cmap='YlOrRd', aspect='auto')
plt.colorbar()
plt.title('Heatmap')

# Treemap
plt.figure()
sizes = treemap_data['Value']
labels = treemap_data['Category']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Treemap')

plt.show()
```

# DISTRIBUTIONS

# Histograms

```python
import matplotlib.pyplot as plt

# Sample data
data = [1, 2, 3, 4, 4, 4, 5, 5, 6, 7, 8, 8, 9]

# Create a histogram
plt.hist(data, bins=5)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [1, 2, 3, 4, 4, 4, 5, 5, 6, 7, 8, 8, 9]

# Calculate mean and variance
mean = np.mean(data)
variance = np.var(data)

# Create a histogram
plt.hist(data, bins=8)

# Add mean and variance lines
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(mean + np.sqrt(variance), color='green', linestyle='dotted', linewidth=2, label='Mean + Variance')
plt.axvline(mean - np.sqrt(variance), color='green', linestyle='dotted', linewidth=2, label='Mean - Variance')

# Add labels, title, and legend
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Mean and Variance')
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
percentages = [390, 240, 410, 120]

# Calculate total value
total_value = sum(percentages)

# Calculate values in whole numbers
values = [(percentage / 100) * total_value for percentage in percentages]
values = [round(value) for value in values]

# Display the values
for i in range(len(categories)):
    print(f"{categories[i]}: {values[i]}")

# Create a pie chart with values
plt.pie(values, labels=categories, autopct='%1.1f%%')

# Add title
plt.title('Pie Chart with Whole Numbers')

# Display the plot
plt.show()
```

# Stacked bar chart

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3']
group1 = [20, 30, 15]
group2 = [10, 25, 20]
group3 = [5, 15, 25]

# Create a stacked bar chart
plt.bar(categories, group1, label='Group 1')
plt.bar(categories, group2, bottom=group1, label='Group 2')
plt.bar(categories, group3, bottom=[group1[j] + group2[j] for j in range(len(group1))], label='Group 3')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')
plt.legend()

# Display the plot
plt.show()
```

# Stacked Area chart

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 15, 25, 30]
y2 = [5, 10, 10, 15, 20]
y3 = [2, 5, 5, 10, 8]

# Create a stacked area chart
plt.stackplot(x, y1, y2, y3, labels=['Group 1', 'Group 2', 'Group 3'])

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stacked Area Chart')
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.random.rand(5, 5)  # 5x5 matrix of random values

# Create a heat map
plt.imshow(data, cmap='hot')

# Add color bar
plt.colorbar()

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heat Map')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.random.rand(5, 5)  # 5x5 matrix of random values

# Create a heat map
plt.imshow(data, cmap='hot')

# Add color bar
plt.colorbar()

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heat Map')

# Customizing tick labels
x_labels = ['A', 'B', 'C', 'D', 'E']
y_labels = ['1', '2', '3', '4', '5']
plt.xticks(range(len(x_labels)), x_labels)
plt.yticks(range(len(y_labels)), y_labels)

# Applying normalization
from matplotlib.colors import Normalize
norm = Normalize(vmin=0, vmax=1)
plt.imshow(data, cmap='hot', norm=norm)

# Adding grid lines
plt.grid(True, linestyle='--', linewidth=0.5)

# Adding annotations
for i in range(len(x_labels)):
    for j in range(len(y_labels)):
        plt.text(i, j, f'{data[j, i]:.2f}', ha='center', va='center', color='white')

# Adjusting figure size
plt.figure(figsize=(8, 6))

# Save the heat map
plt.savefig('heatmap.png')

# Display the plot
plt.show()
```
